# Movie Recommendation System

## Importing the Libraries

```python
import numpy as np
import pandas as pd
import difflib
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
```

## Importing the dataset 

```python
df = pd.read_csv("movies.csv")
df.head()
```

```python
df.shape
```

## Data Preprocessing

```python
selected_columns = ['genres','original_title','keywords','cast','director']
print(selected_columns)
for column in selected_columns:
    df[column]= df[column].fillna("")
```

```python
movies_data = df['genres']+' '+df['original_title']+' ' +df['keywords']+' '+df['cast']+' '+df['director']
movies_data.head()
```

## Converting the text data to Feature Vectors

```python
vectorizer = TfidfVectorizer()
dataset_vector = vectorizer.fit_transform(movies_data)
```

```python
print(dataset_vector)
```

## Compare similarity

```python
similarity = cosine_similarity(dataset_vector)
```

```python
print(similarity)
```

```python
print(similarity.shape)
```

## User Input

```python
movie_name = input("Enter your favourite movie : ")
```

## Search Functionality

```python
movies_list = df["title"].tolist()
```

```python
match = difflib.get_close_matches(movie_name,movies_list)
print(match)
```

```python
index = df[df.title == match[0]]["index"].values[0]
print(index)
```

## Getting similar movies list

```python
similarity_score = list(enumerate(similarity[index]))
print(similarity_score)
                        
```

```python
sorted_similar_movies = sorted(similarity_score, key = lambda x:x[1], reverse=True)
print(sorted_similar_movies)
```

## Printing the Similar movies

```python
print("Movies similar to ",movie_name," : \n")
i = 1
for movie in sorted_similar_movies:
    title = df[df.index == movie[0]]['title'].values[0]
    if(i<30):
        print(i,'. ',title)
        i+=1
```
